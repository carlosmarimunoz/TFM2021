```python
from bs4 import BeautifulSoup
import requests
import re
import pandas as pd
import numpy as np
```

# EXTRACCIÓN DATOS PÁGINA 1 DE IMDB

```python
url = 'https://www.imdb.com/search/title/?companies=co0144901'
response = requests.get(url)
soup = BeautifulSoup(response.text, 'lxml')


li = soup.findAll('a')
lo = li
la = []
for i in range(len(lo)):
        la.append(lo[i].get('href'))


lista_pelis = []
for i in la:
    if "/title" in str(i):
        if "/vote" not in i:
            if "/search" not in i:
                lista_pelis.append(i)
  
lista_pelis = pd.DataFrame(lista_pelis)
lista_pelis = pd.DataFrame(lista_pelis[0].unique())
lista_pelis
```

```python
NETFLIX = pd.DataFrame(columns=['Título','Episodios', 'Rating', 'Votos', 'Géneros', 'Resumen', 'Creador/Director','Num Actores', 'Actores Estrellas','Productoras', 'Pais', 'Localización', 'Num Tags', 'Tags', 'Lenguaje Original', 'Fecha Estreno'])
for i in range(len(lista_pelis[0])):
    url = 'https://www.imdb.com' + lista_pelis[0][i] + '?ref_=adv_li_tt'
    response = requests.get(url)
    soup = BeautifulSoup(response.text, 'lxml')
    title = str(soup.find("meta",  property="og:title").get('content'))
    
    def extraercadena(x):
        indiceinicio = 0
        indicefin = x.find("(")
        textoextraido = x[indiceinicio:(indicefin-1)]
        return textoextraido

    title = [extraercadena(title)]
    title = pd.DataFrame(title)
    title.columns = ['Título']
    

    # EPISODIOS
    
    
    try:
        episodios = soup.find("span",{'class':'bp_sub_heading'}).getText()
        print("Número episodios: " + episodios)
        episodios = [episodios]
        episodios = pd.DataFrame([episodios])
        episodios.columns = ['Episodios']
    except:
        episodios = "No episodes"
        episodios = [episodios]
        episodios = pd.DataFrame([episodios])
        episodios.columns = ['Episodios']
        pass
    

    # RATING & VOTOS
        
    rating_info = soup.find("strong")
    rating = soup.find("strong").getText()
    rating = [rating]
    rating = pd.DataFrame(rating)
    rating.columns = ['Rating']
    
    votes = str(rating_info)[28:-63]
    votes = [votes]
    votes = pd.DataFrame(votes)
    votes.columns = ['Votos']

    hh = soup.findAll("a")

    
    # DIRECTOR
    
    director = []
    for i in hh:
        if "/name/nm" in str(i):
            if len(str(i)) < 100:
                director.append(i.getText())
                
    creador = director[:1]
    creador = [creador]
    creador = pd.DataFrame([creador])
    creador.columns = ['Creador/Director']
       
    # ACTORES
    
    actores = []
    for i in hh:
        if "/name/nm" in str(i):
            if len(str(i)) <100:
                if str(i.getText())[0]==" ":
                       actores.append(str(i.getText())[:-1])
    num_act = len(actores)
    num_act = pd.DataFrame([num_act])
    num_act.columns = ['Num Actores']
    actores = [actores]
    list_act = pd.DataFrame([actores])
    list_act.columns = ['Actores Estrellas']
    

    # PRODUCTORA

    
    productora = []
    for i in hh:
        if "/company/" in str(i):
            productora.append(i.getText())
    productora = [productora]        
    productora = pd.DataFrame([productora])
    productora.columns=['Productoras']
    
    # PAIS

    pais = []
    for i in hh:
        if "/search/title?country" in str(i):
            pais.append(i.getText())
    pais = [pais]
    pais = pd.DataFrame([pais])
    pais.columns=['Pais']




    # LUGAR RODAJE
    
    lugar = []
    for i in hh:
        if "/search/title?location" in str(i):
            if i.getText != "":
                lugar.append(i.getText())
            
    if len(str(lugar)) ==2:
        lugar = [" "]
    
    lugar = [lugar]
    lugar = pd.DataFrame([lugar])
    lugar.columns = ['Localización']
    



    # GENEROS
    
    generos = []
    for i in hh:
        if "/search/title?genres" in str(i):
            if i.getText != "":
                generos.append(i.getText())
                
    generos = [generos]
    generos = pd.DataFrame([generos])
    generos.columns = ['Géneros']
    
    
    
    dvv = soup.find_all("div",{'class':'txt-block'})

    
    # ESTRENO
    
    for i in dvv:
        if "Release" in str(i):
            date = str(i)
                        
    def extraercadena(x):
        indiceinicio = x.find("4>")
        indicefin = x.find("(")
        textoextraido = x[(indiceinicio+3):(indicefin-1)]
        return textoextraido
    date = [extraercadena(date)]
    date = pd.DataFrame([date])
    date.columns = ['Fecha Estreno']



    # LENGUAJE ORIGINAL
    
    for i in dvv:
        if "Langua" in str(i):
            len_ori = i
        
    def extraercadena(x):
        indiceinicio = x.find(":\n")
        textoextraido = x[(indiceinicio+2):-1]
        return textoextraido

    len_ori = extraercadena(len_ori.getText())
    len_ori = [len_ori]
    len_ori = pd.DataFrame([len_ori])
    len_ori.columns = ['Lenguaje Original']
    
    
    # RESUMEN
                       
    data = [d.find_all('span') for d in soup.find_all('p')]

    value = []
    for i in data:
        for j in i:
            value.append(j.text)

    resumen = [(str(value))[6:-2]]
    resumen = pd.DataFrame([resumen])
    resumen.columns = ['Resumen']
    resumen
    


    # PALABRAS CLAVE                
                       
    kw = []
    for i in hh:
        if "/keyword" in str(i):
            kw.append(i)
    link = kw[0].get('href')
    url = 'https://www.imdb.com' + link + '?ref_=tt_stry_kw'

    response = requests.get(url)
    soup = BeautifulSoup(response.text, 'lxml')
    ff = soup.findAll("a")
    ff
    keywords = []
    for i in ff:
        if "/keyword" in str(i):
            keywords.append(i.getText())
        
    num_tags = len(keywords[:-1])
    num_tags = pd.DataFrame([num_tags])
    num_tags.columns = ['Num Tags']
    
    keywords = [keywords[:-1]]
    keywords = pd.DataFrame([keywords])
    keywords.columns = ['Tags']
    




    ## CREACION DATAFRAME POR ROW ##
    
    ROW = pd.concat([title,episodios,rating,votes,generos,resumen,creador,num_act,list_act,productora,pais,lugar,num_tags,keywords,len_ori,date], axis=1)
    a = pd.DataFrame(ROW)
    a['Título'] = a['Título'].astype(str)
    a['Episodios'] = a['Episodios'].astype(str)
    a['Rating'] = a['Rating'].astype(str)
    a['Votos'] = a['Votos'].astype(str)
    a['Géneros'] = a['Géneros'].astype(str)
    a['Resumen'] = a['Resumen'].astype(str)
    a['Creador/Director'] = a['Creador/Director'].astype(str)
    a['Actores Estrellas'] = a['Actores Estrellas'].astype(str)
    a['Productoras'] = a['Productoras'].astype(str)
    a['Pais'] = a['Pais'].astype(str)
    a['Localización'] = a['Localización'].astype(str)
    a['Tags'] = a['Tags'].astype(str)



    ## UNION ##

    NETFLIX= pd.merge(NETFLIX, a, how='outer')
```
